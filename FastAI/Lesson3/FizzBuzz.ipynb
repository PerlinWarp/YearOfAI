# Making our own neuron
A hidden neuron's output can be produced by:
$$ y = \sigma(wx+b) $$
Where:
$$ \sigma(z) =\frac{1}{1+e^{-z}}  $$


```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def hidden_n(w,x,b):
    y = w*x+b
    return sigmoid(y)

hidden_n(1,1,1)
```

```python
# Plotting the output of a hidden neuron 
z = numpy.arange(-5, 5, .1)
sigma_fn = numpy.vectorize(hidden_n)
sigma = sigma_fn(z,1,1)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(z, sigma)
ax.set_ylim([-0.5, 1.5])
ax.set_xlim([-5,5])
ax.grid(True)
ax.set_xlabel('z')
ax.set_title('sigmoid function')

plt.show()
```
