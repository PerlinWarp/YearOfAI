```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
W = np.array([
    [-1,1],
    [1,-1]
])

b = np.array([[-4],
              [-4]])
```

```python
def sigmoid(z):
    # Needed to give non linearity
    return 1/1(1+np.exp(-z))

def neuron(X,W,b):
    # Note that order matters in matrix multiplication. 
    # In these examples, my input space is a point and therefore X is a vector
    return np.dot(X,W) + b
```

```python
X = np.array([
    [1],
    [1]
])

neuron(W, X, b)
```

```python
neuron(1,-1,1)
```

# Looking at NAND from a one neuron network
As the output of a neuron is wX+b, we get a linear line. 
Note as we have two inputs, our space is three dimentional, input x, input y and the output z.

```python
# Plotting the output of a hidden neuron 
x = np.arange(-0.4, 1.5, .05)
y = np.arange(-0.4, 1.5, .05)
sigma_fn = np.vectorize(neuron)
sigma = sigma_fn(x,1,1)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
line, = ax.plot(x, sigma)

# Our inputs we want to give 1 (blue)
positive = np.array([
    [0,0],
    [0,1],
    [1,0],
])
a, b = positive.T
plt.scatter(a,b)

# Our inputs which we want to give 0 (red)
negative = np.array([
    [1,1],
])
a, b= negative.T
plt.scatter(a,b)

ax.set_ylim([-0.5, 1.5])
ax.set_xlim([-0.5,1.5])
ax.grid(True)
ax.set_xlabel('z')
ax.set_title('Sigmoid function')

def update(w = 1.0, b = 1.0):
    line.set_ydata(sigma_fn(x,w,b))
    fig.canvas.draw_idle()

slow = interactive(update,w=FloatSlider(min=-3, max=3, step=0.1), b =FloatSlider(min=-3, max=3, step=0.1));
fig.show()
slow
```

```python
# Predicting using this neuron(W,X,b)
w = -1
b = 1.5
print("A,B,Y")
print(0,0,neuron(0,w,b))
print(0,1,neuron(0,w,b))
print(1,0,neuron(1,w,b))
print(1,1,neuron(1,w,b))
```
